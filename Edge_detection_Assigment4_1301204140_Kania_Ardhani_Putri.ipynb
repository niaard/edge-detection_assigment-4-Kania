Roberts Cross edge detection filter to the image.

```python
#Import Library
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import math
from scipy import ndimage
```

```python
# Load the image
img = cv2.imread('/content/spirited.jpg')

# Check if the image was loaded successfully
if img is not None:
    # Display the image without axis ticks
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Turn off axis ticks
    plt.show()
else:
    print("Failed to load the image. Make sure the image file exists and is supported.")
```

```python
# Definisikan operator Roberts Cross
roberts_cross_v = np.array([[1, 0],
                            [0, -1]])

roberts_cross_h = np.array([[0, 1],
                            [-1, 0]])

# Baca gambar
img = cv2.imread("/content/spirited.jpg", 0)

# Periksa apakah gambar berhasil dibaca
if img is None:
    print("Gagal membaca gambar. Pastikan file gambar ada dan formatnya didukung.")
else:
    # Konversi tipe data ke float64 dan normalisasi
    img = img.astype('float64') / 255.0

    # Terapkan operator Roberts Cross
    vertical = ndimage.convolve(img, roberts_cross_v)
    horizontal = ndimage.convolve(img, roberts_cross_h)

    # Hitung magnitudo gradien
    edged_img = np.sqrt(np.square(horizontal) + np.square(vertical))

    # Normalisasi dan konversi ke rentang 0-255
    edged_img = (edged_img / np.max(edged_img)) * 255

    # Simpan gambar hasil deteksi tepi dalam format JPEG
    output_path = "/content/1301204140_spirited_edges.jpg"
    cv2.imwrite(output_path, edged_img.astype(np.uint8))
    print(f"Output disimpan di: {output_path}")
```

Dengan menggunakan deteksi tepi, gambar hasil menonjolkan batas antara objek dalam gambar. Penyesuaian ukuran juga bermanfaat dalam mengurangi resolusi dan memudahkan pengamatan detail pada gambar yang lebih besar. Keuntungan dari metode ini adalah kemudahan dalam mendeteksi tepi dan orientasi, serta mempertahankan arah diagonal. Namun, metode ini juga memiliki beberapa kelemahan, yaitu sangat sensitif terhadap noise dan tidak begitu akurat dalam mendeteksi tepi. Secara keseluruhan, proses ini dapat memberikan manfaat dalam konteks analisis atau pemrosesan lebih lanjut, tergantung pada kebutuhan spesifiknya.
